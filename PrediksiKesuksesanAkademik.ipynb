# DATA PREPARATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Bar plot for target variable distribution
plt.figure(figsize=(6, 6))
sns.countplot(data=df1, x='Target',edgecolor='black')
plt.title('Distribution of Academic Success')
plt.xlabel('Academic Success')
plt.ylabel('Count')
plt.show();
```

```python
numerical_features = ['Admission grade', 'Age at enrollment', 'Previous qualification (grade)',
                      'Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)',
                      'Unemployment rate', 'Inflation rate', 'GDP']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features):
    plt.subplot(3, 3, i+1)
    sns.boxplot(data=df1, y=feature)
    plt.title(f'{feature} by Academic Success')
    plt.xlabel(' ')
    plt.ylabel(feature)
plt.tight_layout()
plt.show();
```

```python
categorical_features = ['Marital status', 'Application mode', 'Daytime/evening attendance', 
                        'Previous qualification', 'Gender', 'Scholarship holder']

plt.figure(figsize=(14, 14))
for i, feature in enumerate(categorical_features):
    plt.subplot(3, 2, i+1)
    sns.countplot(data=df1, x=feature, hue='Target')
    plt.title(f'{feature} by Academic Success')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.legend(title='Target')
plt.tight_layout()
plt.show();
```

```python
# Calculate the percentage distribution of the target column
target_counts = df1['Target'].value_counts(normalize=True) * 100

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage Distribution of Academic Success')
plt.axis('equal')  
plt.show();
```

```python
df1['Target'] = df1['Target'].replace({'Graduate' : 0,'Dropout' : 1,'Enrolled' : 2})
```

```python
X = df1.drop(columns=['Target','id'])
y = df1['Target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,train_size=0.95,random_state=2)
```

# MODELING

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression(multi_class = 'ovr')
```

```python
logit.fit(X_train,y_train)
```

```python
y_pred_logit = logit.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
cr = classification_report(y_pred_logit,y_test)
print(cr)
```

```python
score = accuracy_score(y_pred_logit,y_test)
print("Accuracy Score of Logistic Regression : %.2f"%(score*100))
print("Error of Logistic Regression : %.2f"%((1 - score)*100))
```
